```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt

fashion_mnist = keras.datasets.fashion_mnist
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

```python
x_train[0]
```

```python
x_train = np.array(x_train, np.float32)
x_test = np.array(x_test, np.float32)
```

```python
x_train[0]
```

```python
num_class = 10
num_features = 28*28

learning_rate = 0.001
training_step = 3000
batch_size = 128
display_step = 100

conv1_filter = 32
conv2_filter = 64

fc1_units = 1024
```

```python
x_train = tf.expand_dims(x_train, axis=-1)
x_test = tf.expand_dims(x_test, axis=-1)
x_train = x_train/255
x_test = x_test/255

train_data = tf.data.Dataset.from_tensor_slices((x_train, y_train))
train_data = train_data.repeat().shuffle(5000).batch(batch_size).prefetch(1)
```

```python
def conv2d(x, W, b, strides = 1):
  x = tf.nn.conv2d(x, W, strides = [1, strides, strides, 1], padding = 'SAME')
  x = tf.nn.bias_add(x,b)
  return tf.nn.relu(x)
```

```python
def maxpool2d(x, k = 2):
  return tf.nn.max_pool(x, ksize = [1, k, k, 1], strides = [1, k, k, 1], padding = 'SAME')
```

```python
random_normal = tf.initializers.RandomNormal()

weights = {'wc1': tf.Variable(random_normal([5,5,1,conv1_filter])),
           'wc2': tf.Variable(random_normal([5,5,conv1_filter,conv2_filter])),
           'wd1': tf.Variable(random_normal([7*7*64, fc1_units])),
           'out': tf.Variable(random_normal([fc1_units,num_class]))}

bias = {'bc1': tf.Variable(tf.zeros(conv1_filter)),
        'bc2': tf.Variable(tf.zeros(conv2_filter)),
        'bd1': tf.Variable(tf.zeros(fc1_units)),
        'out': tf.Variable(tf.zeros(num_class))}
```

```python
def conv_net(x) :
  x = tf.reshape(x, [-1, 28, 28, 1])

  conv1 = conv2d(x, weights['wc1'], bias['bc1'])
  conv1 = maxpool2d(conv1, k = 2)

  conv2 = conv2d(conv1, weights['wc2'], bias['bc2'])
  conv2 = maxpool2d(conv2, k = 2)

  fc1 = tf.reshape(conv2, [-1, weights['wd1'].get_shape().as_list()[0]])
  fc1 = tf.add(tf.matmul(fc1, weights['wd1']), bias['bd1'])

  fc1 = tf.nn.relu(fc1)

  out = tf.add(tf.matmul(fc1, weights['out']), bias['out'])

  return tf.nn.softmax(out)
```

```python
def cross_entropy(y_pred, y_true):
  y_true = tf.one_hot(y_true, depth=num_class)
  y_pred = tf.clip_by_value(y_pred, 1e-9, 1.)
  return tf.reduce_mean(-tf.reduce_sum(y_true * tf.math.log(y_pred), 1))

def accuracy(y_pred, y_true):
  correct_prediction = tf.equal(tf.argmax(y_pred, 1), tf.cast(y_true, tf.int64))
  return tf.reduce_mean(tf.cast(correct_prediction, tf.float32))

optimizer = tf.keras.optimizers.SGD(learning_rate)

def run_optimization(x, y):
  with tf.GradientTape() as g:
    pred = conv_net(x)
    loss = cross_entropy(pred, y)

  trainable_variables = list(weights.values()) + list(bias.values())
  gradients = g.gradient(loss, trainable_variables)
  optimizer.apply_gradients(zip(gradients, trainable_variables))


for step, (batch_x, batch_y) in enumerate(train_data.take(training_step), 1):
  run_optimization(batch_x, batch_y)
  if step % display_step == 0:
    pred = conv_net(batch_x)
    loss = cross_entropy(pred, batch_y)
    acc = accuracy(pred, batch_y)
    print("step: %i, loss: %f, accuracy: %f" % (step, loss, acc))
```

```python
test_pred = conv_net(x_test)
print("Test Accuracy: %f" % accuracy(test_pred, y_test))
```
